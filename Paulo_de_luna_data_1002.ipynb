```python
# Lets also take a look at our crime_index df
crime_index = pd.read_csv('/content/crime_index.csv')
crime_index.tail()
```

```python
# We notice that we have a lot more entraces in crime_index, as opposed to our
# previous dataset. So we are going to merge them based on the countries collumn.

merged_data = pd.merge(data_set, crime_index[['Country', 'Crime Index']], on='Country', how='left')
merged_data.head()
```

```python
# We are going to check for missing values:
merged_data.isnull().sum()
```

```python
len(merged_data)
```

```python
# Lets remove the missing value
merged_data.dropna(inplace=True)
```

```python
#check to see if it's removed
len(merged_data)
```

```python
merged_data.to_csv('/content/merged_countries_data.csv', index=False)
```

```python
# The purpose of our research is to compare Australia's safety record with other
# countries and understand the factors that contribute to its safety.
# By focusing on the selected columns, we ensure that our analysis is targeted
# and relevant. Including only the columns that directly or indirectly influence
# safety allows us to maintain clarity and precision in your findings.
```

```python
# Lets select only the relevant columns for our research
paulo_data_set = merged_data[['Country', 'Safety Index', 'Crime Index', 'Health Care Index', 'Pollution Index', 'Traffic Commute Time Index', 'Climate Index', 'Quality of Life Index', 'Cost of Living Index', 'Purchasing Power Index']]

# Now lets save the cleaned dataset to a new CSV file
paulo_data_set.to_csv('/content/paulo_data_set.csv', index=False)
```
